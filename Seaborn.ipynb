```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
crash_df=sns.load_dataset('car_crashes')
crash_df
```

```python
#distribution plot
#look at univariate distribution
sns.distplot(crash_df['not_distracted'])
```

```python
sns.displot(crash_df['not_distracted'],kde=True)
```

```python
sns.displot(crash_df['not_distracted'],kde=False,bins=25)
```

```python
#joint plot
#compare 2 distributions (a scatter plot by default)

sns.jointplot(x='speeding',y= 'alcohol', data=crash_df)
```

```python
sns.jointplot(x='speeding',y= 'alcohol', data=crash_df, kind='kde') #kernel desnity estimation (estimate distribution of data)
```

```python
sns.jointplot(x='speeding',y= 'alcohol', data=crash_df, kind='kde', fill=True)
```

```python
sns.jointplot(x='speeding',y= 'alcohol', data=crash_df, kind='hex')
```

```python
#Pair plot
#plot the relationship between the entire dataset numerical value

sns.pairplot(crash_df)
```

```python
tips_df=sns.load_dataset('tips')
tips_df
```

```python
sns.pairplot(tips_df, hue='sex', palette='Greens') #hue color by group also use palette shades of a color
```

```python
sns.set_style('darkgrid') #dark, whitegrid, ticks
plt.figure(figsize=(9,4))
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('white') #dark, whitegrid, ticks
plt.figure(figsize=(9,4)) # chhange the whole fig size
sns.set_context('paper', font_scale=2) #increase lable size other option of paper, talk, poster
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
sns.despine(left=True) #remove the left axis other one is bottom top right
```

```python
#categorical plots
#barplots

sns.barplot(x='sex',y='total_bill', data=tips_df)
```

```python
#boxplot
sns.boxplot(x='day',y='total_bill',data=tips_df, hue='sex')
```

```python
#violinplot
sns.violinplot(x='day',y='total_bill',data=tips_df, hue='sex')
```

```python
#stripplot
sns.stripplot(x='day',y='total_bill',data=tips_df, hue='sex', jitter=True)#to minimise data point overlap
```

```python
sns.violinplot(x='day',y='total_bill',data=tips_df)
sns.swarmplot(x='day',y='total_bill',data=tips_df,color='white')
```

```python
crash_mx=crash_df.corr()
crash_mx
```

```python
sns.heatmap(crash_mx,annot=True, cmap='summer')
```
