```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="ticks", color_codes=True)
iris = sns.load_dataset("iris")
g = sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris, hue="species")
```

```python
sns.pairplot(iris, kind="kde")
```

```python
g = sns.pairplot(iris, vars=["sepal_width", "sepal_length"])
```

```python
g = sns.pairplot(iris, diag_kind="kde")
```

```python
#Color Qualitative Palatte
from matplotlib import pyplot as pyplot
import seaborn as sns 
new_palette=sns.color_palette()
sns.palplot(new_palette)
plt.show()
```

```python
#Sequential Color Palette
new_palette=sns.color_palette("pastel")
sns.palplot(new_palette)
plt.show()
```

```python
#Diverging color palette
new1_palette=sns.color_palette("BrBG",8)
sns.palplot(new1_palette)
plt.show()
```

```python
#Lineplot
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("iris")
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("iris")
sns.barplot(x="sepal_length", y="sepal_width", data=data)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("iris")
sns.boxplot(x="sepal_length", y="sepal_width", data=data)
plt.show()
```

```python
#Pairplot
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
sns.countplot(x ='sepal_width', data = df)
plt.show()
```

```python
#Hexbinplot
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('iris')
sb.jointplot(x = 'petal_length',y = 'petal_width',data = df,kind = 'hex')
plt.show()
```

```python
#VioinPlot
import seaborn
seaborn.set(style = 'whitegrid')
fmri = seaborn.load_dataset("iris")
seaborn.violinplot(x ="sepal_width",
y ="petal_width",
data = iris)
```

```python
#Stripplot
import seaborn
import matplotlib.pyplot as plt
seaborn.set(style = 'whitegrid')  
tip = seaborn.load_dataset("iris")  
seaborn.stripplot(x="sepal_width", y="petal_width", data=tip)
plt.show()
```
